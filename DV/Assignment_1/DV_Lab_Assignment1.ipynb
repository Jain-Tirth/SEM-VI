## 1. Box Plot
- Create a box plot to compare the distribution of math scores across different gender categories.
- Show median, quartiles, and outliers clearly.
- Title: "Math Score Distribution by Gender".
- Label the x-axis as "Gender" and the y-axis as "Math Score".

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/StudentsPerformance.csv")
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=df,
    x="gender",
    y="math score",
    palette="Set2"
)

plt.title("Math Score Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")

plt.savefig("Q1_BoxPlot.png")
plt.show()
```

## 2. Violin Plot
- Create a violin plot to visualize the distribution of writing scores for different parental level of
education.
- Use different colors for each education level.
- Title: "Writing Score Distribution by Parental Education".
- Label the x-axis as "Parental Level of Education" and the y-axis as "Writing Score".

```python
plt.figure(figsize=(10,6))

sns.violinplot(
    data=df,
    x="parental level of education",
    y="writing score",
    palette="Set3"
)

plt.title("Writing Score Distribution by Parental Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Writing Score")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("Q2_ViolinPlot.png")
plt.show()
```

## 3. Heatmap
- Create a heatmap to display the correlation matrix of math, reading, and writing scores.
- Use a colorbar to represent correlation strength.
- Annotate each cell with correlation values.
- Title: "Correlation Heatmap of Student Scores".

```python
corr = df[["math score", "reading score", "writing score"]].corr()

plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    cbar=True
)

plt.title("Correlation Heatmap of Student Scores")

plt.savefig("Q3_Heatmap.png")
plt.show()
```

## 4. Subplots (Dashboard View)
- Create a 2×2 subplot layout containing:
- Histogram of math scores.
- Box plot of reading scores by gender.
- Scatter plot of math vs writing scores.
- Bar chart of average writing score by gender.
- Add a common figure title.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].hist(df["math score"], bins=15, color="skyblue", edgecolor="black")
axes[0, 0].set_title("Histogram of Math Scores")

sns.boxplot(data=df, x="gender", y="reading score", ax=axes[0, 1], palette="Set2")
axes[0, 1].set_title("Reading Score by Gender")

axes[1, 0].scatter(df["math score"], df["writing score"], color="red", alpha=0.6)
axes[1, 0].set_title("Math vs Writing Scores")
axes[1, 0].set_xlabel("Math Score")
axes[1, 0].set_ylabel("Writing Score")

avg_writing = df.groupby("gender")["writing score"].mean()
axes[1, 1].bar(avg_writing.index, avg_writing.values, color=["steelblue", "salmon"])
axes[1, 1].set_title("Average Writing Score by Gender")

fig.suptitle("Student Performance Dashboard", fontsize=16)

plt.tight_layout()
plt.savefig("Q4_Subplots.png")
plt.show()
```

## 5. Draw a house diagram using matplotlib.patches

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(6,6))

ax.set_xlim(0,7)
ax.set_ylim(0,7)

house_body = patches.Rectangle((1,1), 4, 3, facecolor="#e6e6c8", edgecolor="black")
ax.add_patch(house_body)

roof = patches.Polygon([[1,4], [3,6], [5,4]], facecolor="darkred", edgecolor="black")
ax.add_patch(roof)

chimney = patches.Rectangle((3.8,4.5), 0.4, 1, facecolor="darkred", edgecolor="black")
ax.add_patch(chimney)

door = patches.Rectangle((2.5,1), 1, 1.6, facecolor="saddlebrown", edgecolor="black")
ax.add_patch(door)

door_knob = patches.Circle((3.3,1.8), 0.1, facecolor="gold", edgecolor="black")
ax.add_patch(door_knob)

window1 = patches.Rectangle((1.6,2.3), 0.7, 0.6, facecolor="#66b3cc", edgecolor="black")
ax.add_patch(window1)

window2 = patches.Rectangle((3.7,2.3), 0.7, 0.6, facecolor="#66b3cc", edgecolor="black")
ax.add_patch(window2)

sun = patches.Circle((6,6), 0.5, facecolor="orange", edgecolor="black")
ax.add_patch(sun)

ground = patches.Rectangle((0,0.85), 7, 0.1, facecolor="green")
ax.add_patch(ground)

ax.set_title("House Diagram using matplotlib.patches")
ax.grid(True)
ax.set_aspect("equal")

plt.savefig("Q5_HouseDiagram.png")
plt.show()
```

## 6. Generate a 3D surface plot for the mathematical function:𝑧 = 𝑥2 + 𝑦2
## Use proper axis labels and title.

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap="viridis")

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Surface Plot of z = x² + y²")

plt.savefig("Q6_3DSurface.png")
plt.show()
```

## 7. Using the matplotlib.image sub-library, perform the following operations on a single input image and display all outputs appropriately:

- Load and Display Image
  - a. Read the image using matplotlib.image.imread().
  - b. Display the original image.

- Convert to Grayscale
  - c. Convert the loaded image into grayscale format.
  - d. Display both original and grayscale images using subplots.
- Image Manipulation
  - e. Crop a selected region of the image.
  - f. Flip the image horizontally.

  - g. Display original, cropped, and flipped images in a single figure.Brightness and Contrast Adjustment

  - h. Increase the brightness of the image.
  - i. Modify the contrast of the image.
  - j. Display original and modified images side by side for comparision.
  
- Watermarking
   - k. Overlay a transparent watermark text on the original image using Matplotlib functions.
  - l. The watermark should appear semi-transparent and positioned at a visible location.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread("/content/input.png")

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.savefig("Q7_Original.png")
plt.show()

gray = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(gray, cmap="gray")
ax[1].set_title("Grayscale")
ax[1].axis("off")

plt.savefig("Q7_Grayscale.png")
plt.show()

cropped = img[100:400, 100:400]

flipped = np.fliplr(img)

fig, ax = plt.subplots(1,3, figsize=(12,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(cropped)
ax[1].set_title("Cropped")
ax[1].axis("off")

ax[2].imshow(flipped)
ax[2].set_title("Flipped")
ax[2].axis("off")

plt.savefig("Q7_Crop_Flip.png")
plt.show()

bright = np.clip(img + 0.2, 0, 1)

contrast = np.clip((img - 0.5) * 1.5 + 0.5, 0, 1)

fig, ax = plt.subplots(1,3, figsize=(12,5))
ax[0].imshow(img)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(bright)
ax[1].set_title("Brightness Increased")
ax[1].axis("off")

ax[2].imshow(contrast)
ax[2].set_title("Contrast Adjusted")
ax[2].axis("off")

plt.savefig("Q7_Brightness_Contrast.png")
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.text(
    img.shape[1]//3,
    img.shape[0]//2,
    "WATERMARK",
    fontsize=40,
    color="white",
    alpha=0.4
)
plt.axis("off")
plt.savefig("Q7_Watermark.png")
plt.show()
```
